```python
#importing librabries
import pandas as pd

#Reading the text file
df = pd.read_csv('showwcase.csv')
df.head()
```

```python
#Checking the datatypes
df.dtypes
```

```python
#Converting login_date to a datetype
df['login_date'] = pd.to_datetime(df['login_date'],format = '%m/%d/%y')
```

```python
#Pre-processing Data
#Cross verifying the null value to ensure its not part of the original dataset
df.isnull().sum()
```

```python
#Deleting the null values of the dataset occured due to formatting issues
df.dropna()
```

```python
# Dropping rows wherein inactive duration is greater than session duration 
df=df[df['inactive_duration']<df['session_duration']]  # According to definition session duration is when the user is logged for the session 
```

```python
#Adding new columns 
df['day_of_week'] = df['login_date'].dt.day_name()  #Day_of_the week 
df['inactive_duration_min'] = (df['inactive_duration']/60 ) .round(2)  #Converting  inactive_duration to minutes
df['session_duration_min'] = (df['session_duration']/60 ) .round(2)    #Coverting to session_duration to minutes
df.head(10)
```
